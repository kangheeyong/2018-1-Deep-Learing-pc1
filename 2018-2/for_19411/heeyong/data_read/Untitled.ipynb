```python

import ipyvolume as ipv
import numpy as np

x = y= np.arange(0, 128)
X, Y = np.meshgrid(x, y)
zs = np.array([128*(normal_data[1110,x,y]-3000)/7000 for x,y in zip(np.ravel(X), np.ravel(Y))])
Z = zs.reshape(X.shape)

ipv.figure()

m = ipv.plot_mesh(X, Y, Z, wireframe=False)

ipv.xlabel('time')
ipv.ylabel('wavelength')
ipv.zlabel('intensity')
ipv.show()
```

```python
import ipyvolume as ipv
import numpy as np


a = np.zeros((128,128,128))



a.shape
a[0,0,0]
```

```python
ipv.clear()

x = []
y = []
z = []
for i in range(128) :
    for j in range(128) : 
        a[i,j,int(128*(normal_data[1110,i,j]-3000)/7000)] = 1
        x.append(i)
        y.append(j)
        z.append(int(128*(normal_data[1110,i,j]-3000)/7000))
x = np.array(x)/128
y = np.array(y)/128
z = np.array(z)/128

ipv.scatter(x, y, z, marker="box",size=1)
ipv.style.box_on()
ipv.show()
```

```python
ipv.clear()

x = []
y = []
z = []
for i in range(128) :
    for j in range(128) : 
        
        temp = np.maximum(0,int(128*(anomaly_data[110,i,j]-3000)/7000))
        temp = np.minimum(127,int(128*(anomaly_data[110,i,j]-3000)/7000))
        
        a[i,j,temp] = 1
        x.append(i)
        y.append(j)
        z.append(temp)
x = np.array(x)/128
y = np.array(y)/128
z = np.array(z)/128
ipv.figure()
ipv.quickscatter(x, y, z, marker="box",size=1,color=(1,1,0))
ipv.show()

ipv.pylab.save('sdf.html')
```
