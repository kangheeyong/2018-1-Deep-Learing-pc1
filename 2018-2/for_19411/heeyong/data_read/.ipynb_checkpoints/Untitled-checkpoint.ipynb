```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
import gzip

```

```python
dir_file = 'normal/'
with gzip.open(dir_file  + 'normal_data.pickle.gzip','rb') as f :
    normal_data = pickle.load(f)
```

```python
normal_data.shape
```

```python
dir_file = 'anomaly/'
with gzip.open(dir_file  + 'anomaly_data.pickle.gzip','rb') as f :
    anomaly_data = np.array(pickle.load(f))
```

```python
anomaly_data.shape
```

```python

import ipyvolume as ipv
import numpy as np

x = y= np.arange(0, 128)
X, Y = np.meshgrid(x, y)
zs = np.array([normal_data[1110,x,y] for x,y in zip(np.ravel(X), np.ravel(Y))])
Z = zs.reshape(X.shape)

ipv.figure()
m = ipv.plot_mesh(X, Y, Z, wireframe=False)

ipv.pylab.xlim(0, 128)
ipv.pylab.ylim(0, 128)
#ipv.pylab.zlim(7, 9)
ipv.show()
```

```python

import ipyvolume as ipv
import numpy as np

x = y= np.arange(0, 128)
X, Y = np.meshgrid(x, y)
zs = np.array([anomaly_data[10,x,y] for x,y in zip(np.ravel(X), np.ravel(Y))])
Z = zs.reshape(X.shape)

ipv.figure()
m = ipv.plot_mesh(X, Y, Z, wireframe=False)

ipv.pylab.xlim(0, 128)
ipv.pylab.ylim(0, 128)
#ipv.pylab.zlim(7, 9)
ipv.show()

ipv.pylab.save('sdf.html')
```
