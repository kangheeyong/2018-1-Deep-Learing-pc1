# Define Parameters of my NN Model

## 1st

```python
learning_rate = 0.00001

n_InputUnit = 4
n_1stHiddenUnit=18
n_2ndHiddenUnit=18
n_3rdHiddenUnit=9
n_4thHiddenUnit=9
n_OutputUnit = 1

W1 = tf.Variable(tf.random_normal([n_InputUnit, n_1stHiddenUnit],stddev=0.1))
W2 = tf.Variable(tf.random_normal([n_1stHiddenUnit, n_2ndHiddenUnit],stddev=0.1))
W3 = tf.Variable(tf.random_normal([n_2ndHiddenUnit, n_3rdHiddenUnit],stddev=0.1))
W4 = tf.Variable(tf.random_normal([n_3rdHiddenUnit, n_4thHiddenUnit],stddev=0.1))
W5 = tf.Variable(tf.random_normal([n_4thHiddenUnit, n_OutputUnit],stddev=0.1))

b1 = tf.Variable(tf.random_normal([n_1stHiddenUnit], mean = 0, stddev=0.01))
b2 = tf.Variable(tf.random_normal([n_2ndHiddenUnit], mean = 0, stddev=0.01))
b3 = tf.Variable(tf.random_normal([n_3rdHiddenUnit], mean = 0, stddev=0.01))
b4 = tf.Variable(tf.random_normal([n_4thHiddenUnit], mean = 0, stddev=0.01))
b5 = tf.Variable(tf.random_normal([n_OutputUnit], mean = 0, stddev=0.01))

X = tf.placeholder(tf.float32, [None, 4]) # tensorflow의 placeholder라는 변수를 선언
Y = tf.placeholder(tf.float32, [None,])

L1 = tf.nn.relu(tf.matmul(X, W1) + b1)
L2 = tf.nn.relu(tf.matmul(L1, W2) + b2)
L3 = tf.nn.relu(tf.matmul(L2, W3) + b3)
L4 = tf.nn.relu(tf.matmul(L3, W4) + b4)
Hypothesis = tf.nn.relu(tf.matmul(L4, W5) + b5)

cost = tf.reduce_mean((Y-Hypothesis)*(Y-Hypothesis))
train = tf.train.GradientDescentOptimizer(learning_rate=learning_rate).minimize(cost)
```

```python
tf.set_random_seed(777)  # for reproducibility
learning_rate = 0.00001

n_InputUnit = 4
n_1stHiddenUnit=20
n_2ndHiddenUnit=20
n_3rdHiddenUnit=20
n_4thHiddenUnit=20
n_5thHiddenUnit=20
n_6thHiddenUnit=20
n_7thHiddenUnit=20
n_OutputUnit = 1

W1 = tf.Variable(tf.random_normal([n_InputUnit, n_1stHiddenUnit],stddev=0.01))
W2 = tf.Variable(tf.random_normal([n_1stHiddenUnit, n_2ndHiddenUnit],stddev=0.01))
W3 = tf.Variable(tf.random_normal([n_2ndHiddenUnit, n_3rdHiddenUnit],stddev=0.01))
W4 = tf.Variable(tf.random_normal([n_3rdHiddenUnit, n_4thHiddenUnit],stddev=0.01))
W5 = tf.Variable(tf.random_normal([n_4thHiddenUnit, n_5thHiddenUnit],stddev=0.01))
W6 = tf.Variable(tf.random_normal([n_5thHiddenUnit, n_6thHiddenUnit],stddev=0.01))
W7 = tf.Variable(tf.random_normal([n_6thHiddenUnit, n_7thHiddenUnit],stddev=0.01))
W8 = tf.Variable(tf.random_normal([n_7thHiddenUnit, n_OutputUnit],stddev=0.01))

b1 = tf.Variable(tf.random_normal([n_1stHiddenUnit], stddev=0.01))
b2 = tf.Variable(tf.random_normal([n_2ndHiddenUnit], stddev=0.01))
b3 = tf.Variable(tf.random_normal([n_3rdHiddenUnit], stddev=0.01))
b4 = tf.Variable(tf.random_normal([n_4thHiddenUnit], stddev=0.01))
b5 = tf.Variable(tf.random_normal([n_5thHiddenUnit], stddev=0.01))
b6 = tf.Variable(tf.random_normal([n_6thHiddenUnit], stddev=0.01))
b7 = tf.Variable(tf.random_normal([n_7thHiddenUnit], stddev=0.01))
b8 = tf.Variable(tf.random_normal([n_OutputUnit], stddev=0.01))

X = tf.placeholder(tf.float32, [None, 4]) # tensorflow의 placeholder라는 변수를 선언
Y = tf.placeholder(tf.float32, [None,])

L1 = tf.nn.relu(tf.matmul(X, W1) + b1)
L2 = tf.nn.relu(tf.matmul(L1, W2) + b2)
L3 = tf.nn.relu(tf.matmul(L2, W3) + b3)
L4 = tf.nn.relu(tf.matmul(L3, W4) + b4)
L5 = tf.nn.relu(tf.matmul(L4, W5) + b5)
L6 = tf.nn.relu(tf.matmul(L5, W6) + b6)
L7 = tf.nn.relu(tf.matmul(L6, W7) + b7)
Hypothesis = tf.nn.tanh(tf.matmul(L7, W8) + b8)

cost = tf.reduce_mean((Y-Hypothesis)*(Y-Hypothesis))
train = tf.train.AdamOptimizer(learning_rate, beta1 = 0.01, beta2 = 0.1).minimize(cost)#######
```

# Train NN & Predict

```python
with tf.Session() as session:
    session.run(tf.global_variables_initializer())
    
    for step in range(300000):
        SGDorder = Sequence[step % 21] # training을 0번부터 23번까지 순차적으로 돌린다.
        session.run(train, feed_dict ={X: X_TrainingData[SGDorder: SGDorder + 1,:], Y: Y_TrainingData[SGDorder: SGDorder + 1]})
        #session.run(train, feed_dict ={X: X_TrainingData, Y: Y_TrainingData})
        
        if step % 1000 == 0:
            print (SGDorder,step, session.run(cost, feed_dict = {X: X_TrainingData, Y: Y_TrainingData}))
            #print (W1)
            #print (b1)
        if step % 21 == 0:
            random.shuffle(Sequence)
            
    Predict_to_TestData = session.run(Hypothesis, feed_dict={X : X_TestData})
    Predict_to_TrainingData = session.run(Hypothesis, feed_dict={X : X_TrainingData})
    Predict_to_EntireData = session.run(Hypothesis, feed_dict = {X : X_NormalizedData})
    
    print("R_square")
    print(r2_score(Y_NormalizedData, Predict_to_EntireData))
    print("Result of predicting TrainingData")
    print((Predict_to_TrainingData * (max(Y_OriginalData)-min(Y_OriginalData))) + np.mean(Y_OriginalData))
    print("Result of predicting TestData")
    print((Predict_to_TestData * (max(Y_OriginalData)-min(Y_OriginalData))) + np.mean(Y_OriginalData))
    print("Original Data")
    print(Y_OriginalData)
```
