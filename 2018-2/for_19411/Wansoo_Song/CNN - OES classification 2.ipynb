```python
learning_rate = 0.00005
tf.set_random_seed(777)
batchsize = 10

def build_CNN_classifier(X):
    
    W_conv1 = tf.Variable(tf.truncated_normal(shape=[1, 1, 57, 57], stddev=0.01)) #filter
    # 1 * 1 * image depth * filter 2개
    b_conv1 = tf.Variable(tf.constant(0.1, shape=[57])) #bias
    # convolution filter 갯수에 맞춤 = 2개
    h_conv1 = tf.nn.elu(tf.nn.conv2d(X, W_conv1, strides=[1, 1, 1, 1], padding='VALID') + b_conv1)
    print(h_conv1)
    # convolution 연산 수행. 
    # stride >> 2칸씩 움직이는게 목표라면 [1, 2, 2, 1]로 수정
    # padding >> 'SAME' 이면 stride에 맞추어 최소 size의 padding으로 convolution
    #            'VALID' 이면 stride에 맞추어 계산 가능한 만큼만 계산. 나머지 계산 안된건 버림
    
  # 첫번째 Pooling layer
  #h_pool1 = tf.nn.max_pool(h_conv1, ksize=[1, 3, 3, 1], strides=[1, 2, 2, 1], padding='SAME')
    # 3*3 필터를 2의 stride를 가지고 움직임
    # max pooling

  # 2nd : 8*8*57 -> 5*5*57*114(filter) -> 4*4*114
    W_conv2 = tf.Variable(tf.truncated_normal(shape=[5, 5, 57, 114], stddev=0.01))
    b_conv2 = tf.Variable(tf.constant(0.1, shape=[114]))
    h_conv2 = tf.nn.elu(tf.nn.conv2d(h_conv1, W_conv2, strides=[1, 1, 1, 1], padding='VALID') + b_conv2)
    print(h_conv2)

  # 3rd : 4*4*114 -> 3*3*114*38(filter) -> 2*2*38
    W_conv3 = tf.Variable(tf.truncated_normal(shape=[3, 3, 114, 38], stddev=0.01))
    b_conv3 = tf.Variable(tf.constant(0.1, shape=[38]))
    h_conv3 = tf.nn.elu(tf.nn.conv2d(h_conv2, W_conv3, strides=[1, 1, 1, 1], padding='VALID') + b_conv3)
    print(h_conv3)

  # 2*2*38 -> 8개의 neuron으로 (FCNN)
    W_fc1 = tf.Variable(tf.truncated_normal(shape=[2*2*38, 8], stddev=0.01))
    b_fc1 = tf.Variable(tf.constant(0.1, shape=[8]))
    h_conv4_flat = tf.reshape(h_conv3, [-1, 2*2*38])
    h_fc1 = tf.nn.elu(tf.matmul(h_conv4_flat, W_fc1) + b_fc1)
    print(h_fc1)
    
  # 8 -> 4
    W_fc2 = tf.Variable(tf.truncated_normal(shape=[8, 4], stddev=0.01))
    b_fc2 = tf.Variable(tf.constant(0.1, shape=[4]))
    y_pred = tf.nn.elu(tf.matmul(h_fc1,W_fc2) + b_fc2)
    print(y_pred)

    return y_pred

X = tf.placeholder(tf.float32, shape=[None, 8, 8, 57])
Y = tf.placeholder(tf.float32, shape=[None, 4])

y_pred = build_CNN_classifier(X)

cost = tf.reduce_mean(tf.pow(Y - y_pred, 2))
train = tf.train.AdamOptimizer(learning_rate=learning_rate, beta1 = 0.5).minimize(cost)

Accuracy = y_pred
```

```python
learning_rate = 0.0001
tf.set_random_seed(777)
batchsize = 10

def build_CNN_classifier(X):
    
    W_conv1 = tf.Variable(tf.truncated_normal(shape=[1, 1, 57, 57], stddev=0.01)) #filter
    # 1 * 1 * image depth * filter 2개
    b_conv1 = tf.Variable(tf.constant(0.1, shape=[57])) #bias
    # convolution filter 갯수에 맞춤 = 2개
    h_conv1 = tf.nn.elu(tf.nn.conv2d(X, W_conv1, strides=[1, 1, 1, 1], padding='VALID') + b_conv1)
    print(h_conv1)
    # convolution 연산 수행. 
    # stride >> 2칸씩 움직이는게 목표라면 [1, 2, 2, 1]로 수정
    # padding >> 'SAME' 이면 stride에 맞추어 최소 size의 padding으로 convolution
    #            'VALID' 이면 stride에 맞추어 계산 가능한 만큼만 계산. 나머지 계산 안된건 버림
    
  # 첫번째 Pooling layer
  #h_pool1 = tf.nn.max_pool(h_conv1, ksize=[1, 3, 3, 1], strides=[1, 2, 2, 1], padding='SAME')
    # 3*3 필터를 2의 stride를 가지고 움직임
    # max pooling

  # 2nd : 8*8*57 -> 5*5*57*114(filter) -> 4*4*57
    W_conv2 = tf.Variable(tf.truncated_normal(shape=[5, 5, 57, 57], stddev=0.01))
    b_conv2 = tf.Variable(tf.constant(0.1, shape=[57]))
    h_conv2 = tf.nn.elu(tf.nn.conv2d(h_conv1, W_conv2, strides=[1, 1, 1, 1], padding='VALID') + b_conv2)
    print(h_conv2)

  # 3rd : 4*4*57 -> 3*3*57*19(filter) -> 2*2*19
    W_conv3 = tf.Variable(tf.truncated_normal(shape=[3, 3, 57, 19], stddev=0.01))
    b_conv3 = tf.Variable(tf.constant(0.1, shape=[19]))
    h_conv3 = tf.nn.elu(tf.nn.conv2d(h_conv2, W_conv3, strides=[1, 1, 1, 1], padding='VALID') + b_conv3)
    print(h_conv3)

  # 2*2*38 -> 8개의 neuron으로 (FCNN)
    W_fc1 = tf.Variable(tf.truncated_normal(shape=[2*2*19, 8], stddev=0.01))
    b_fc1 = tf.Variable(tf.constant(0.1, shape=[8]))
    h_conv4_flat = tf.reshape(h_conv3, [-1, 2*2*19])
    h_fc1 = tf.nn.elu(tf.matmul(h_conv4_flat, W_fc1) + b_fc1)
    print(h_fc1)
    
  # 8 -> 4
    W_fc2 = tf.Variable(tf.truncated_normal(shape=[8, 4], stddev=0.01))
    b_fc2 = tf.Variable(tf.constant(0.1, shape=[4]))
    y_pred = tf.nn.elu(tf.matmul(h_fc1,W_fc2) + b_fc2)
    print(y_pred)

    return y_pred

X = tf.placeholder(tf.float32, shape=[None, 8, 8, 57])
Y = tf.placeholder(tf.float32, shape=[None, 4])

y_pred = build_CNN_classifier(X)

cost = tf.reduce_mean(tf.pow(Y - y_pred, 2))
train = tf.train.AdamOptimizer(learning_rate=learning_rate).minimize(cost)

Accuracy = y_pred
```

```python
learning_rate = 0.0001
tf.set_random_seed(777)
batchsize = 10

def build_CNN_classifier(X):
    
    W_conv1 = tf.Variable(tf.truncated_normal(shape=[1, 1, 57, 57], stddev=0.01)) #filter
    # 1 * 1 * image depth * filter 57개
    b_conv1 = tf.Variable(tf.constant(0.1, shape=[57])) #bias
    # convolution filter 갯수에 맞춤 = 2개
    h_conv1 = tf.nn.elu(tf.nn.conv2d(X, W_conv1, strides=[1, 1, 1, 1], padding='VALID') + b_conv1)
    
    print(h_conv1)
    # convolution 연산 수행. 
    # stride >> 2칸씩 움직이는게 목표라면 [1, 2, 2, 1]로 수정
    # padding >> 'SAME' input image의 크기와 똑같이 계산
    #            'VALID' 이면 stride에 맞추어 계산 가능한 만큼만 계산. 나머지 계산 안된건 버림
    
  # 첫번째 Pooling layer
  #h_pool1 = tf.nn.max_pool(h_conv1, ksize=[1, 3, 3, 1], strides=[1, 2, 2, 1], padding='SAME')
    # 3*3 필터를 2의 stride를 가지고 움직임
    # max pooling

  # 2nd : 8*8*57 -> 5*5*57*19(filter) -> 4*4*19 -> 2*2*19
    W_conv2 = tf.Variable(tf.truncated_normal(shape=[5, 5, 57, 19], stddev=0.01))
    b_conv2 = tf.Variable(tf.constant(0.1, shape=[19]))
    h_conv2 = tf.nn.elu(tf.nn.conv2d(h_conv1, W_conv2, strides=[1, 1, 1, 1], padding='VALID') + b_conv2)
    h_pool1 = tf.nn.max_pool(h_conv2, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='VALID')
    print(h_conv2)
    print(h_pool1)
    
  # 3rd : 2*2*19 -> 3*3*19*5(filter) -> 2*2*5
    W_conv3 = tf.Variable(tf.truncated_normal(shape=[3, 3, 19, 5], stddev=0.01))
    b_conv3 = tf.Variable(tf.constant(0.1, shape=[5]))
    h_conv3 = tf.nn.elu(tf.nn.conv2d(h_pool1, W_conv3, strides=[1, 1, 1, 1], padding='SAME') + b_conv3)
    print(h_conv3)

  # 2*2*5 -> 8개의 neuron으로 (FCNN)
    W_fc1 = tf.Variable(tf.truncated_normal(shape=[2*2*5, 8], stddev=0.01))
    b_fc1 = tf.Variable(tf.constant(0.1, shape=[8]))
    h_conv4_flat = tf.reshape(h_conv3, [-1, 2*2*5])
    h_fc1 = tf.nn.elu(tf.matmul(h_conv4_flat, W_fc1) + b_fc1)
    print(h_fc1)
    
  # 8 -> 4
    W_fc2 = tf.Variable(tf.truncated_normal(shape=[8, 4], stddev=0.01))
    b_fc2 = tf.Variable(tf.constant(0.1, shape=[4]))
    y_pred = tf.nn.elu(tf.matmul(h_fc1,W_fc2) + b_fc2)
    print(y_pred)

    return y_pred

X = tf.placeholder(tf.float32, shape=[None, 8, 8, 57])
Y = tf.placeholder(tf.float32, shape=[None, 4])

y_pred = build_CNN_classifier(X)

cost = tf.reduce_mean(tf.pow(Y - y_pred, 2))
train = tf.train.AdamOptimizer(learning_rate=learning_rate, beta1 = 0.5).minimize(cost)

Accuracy = y_pred
```
