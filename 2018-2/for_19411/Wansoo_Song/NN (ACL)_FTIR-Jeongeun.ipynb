# 정은이가 준 코드 (4 30 30 30 1)

```python
import numpy as np
import tensorflow as tf
from sklearn.metrics import r2_score
import os

os.environ["CUDA_VISIBLE_DEVICES"]="-1"

data = np.loadtxt("ACL FTIR data.txt")

X_data = data[:,0:4]
Y_data = data[:,4]

X_train = data[0:20,0:4]
Y_train = data[0:20,4]

X_test = data[20:,0:4]
Y_test = data[20:,4]

tf.set_random_seed(777)  # for reproducibility
learning_rate = 0.001

X = tf.placeholder(tf.float32, [None, 4])
Y = tf.placeholder(tf.float32, [None,])

W1 = tf.Variable(tf.random_normal([4, 30],stddev=0.01), name='weight1')
b1 = tf.Variable(tf.random_normal([30], stddev=0.01), name='bias1')
layer1 = tf.nn.elu(tf.matmul(X, W1) + b1)

W2 = tf.Variable(tf.random_normal([30, 30],stddev=0.01), name='weight2')
b2 = tf.Variable(tf.random_normal([30], stddev=0.01), name='bias2')
layer2 = tf.nn.elu(tf.matmul(layer1, W2) + b2)

W3 = tf.Variable(tf.random_normal([30, 30],stddev=0.01), name='weight3')
b3 = tf.Variable(tf.random_normal([30], stddev=0.01), name='bias3')
layer3 = tf.nn.elu(tf.matmul(layer2, W3) + b3)

W4 = tf.Variable(tf.random_normal([30, 1],stddev = 0.01), name='weight4')
b4 = tf.Variable(tf.random_normal([1], stddev=0.01), name='bias4')
hypothesis = tf.nn.sigmoid(tf.matmul(layer3, W4) + b4)

# cost/loss function
cost = tf.reduce_mean(tf.square(Y-hypothesis))

train = tf.train.AdamOptimizer(learning_rate=learning_rate,beta1=0.5).minimize(cost)

with tf.Session() as sess:
    # Initialize TensorFlow variables
    sess.run(tf.global_variables_initializer())
            
    for step in range(80000):
        idx = step % 20
        sess.run(train, feed_dict={X: X_train[idx: idx + 1,:], Y: Y_train[idx:idx+1]})
        if step % 1000 == 0:
            print(step, sess.run(cost, feed_dict={X: X_train, Y: Y_train})) 
                                                        
                                                            
                                                                                    
    result = sess.run(hypothesis, feed_dict={X : X_test})
    result_train = sess.run(hypothesis, feed_dict={X : X_train})
    result_data = sess.run(hypothesis, feed_dict = {X : X_data})
    
    print("result")
    print(result)
    print("Y_test")
    print(Y_test)
    print("result_train")
    print(result_train)
    print("result_data")
    print(result_data)
    
    real = Y_data
    prediction = result_data
    
    print("real")
    print(real)
    print("prediction")
    print(prediction)
    print("R square")
    print(r2_score(real,prediction))
    print("R square_Test")
    print(r2_score(Y_test,result))
```
