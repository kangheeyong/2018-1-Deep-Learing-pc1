```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import statsmodels.formula.api as sm
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn import cross_validation
import tensorflow as tf
```

```python
data = np.genfromtxt('acl data.txt')

X_before = data[:,0:4]
Y_before = data[:,4]

X = (X_before-np.min(X_before))/(np.max(X_before)-np.min(X_before))
Y = (Y_before-min(Y_before))/(max(Y_before)-min(Y_before))   #nomalizing

dfX=pd.DataFrame(X)
dfy=pd.DataFrame(Y)

table=pd.concat([dfX, dfy], axis=1)
table.columns=['Power','Pressure','C3H6','N2','Thickness_Top']

table
```

```python
X_train, X_test, Y_train, Y_test = cross_validation.train_test_split(
X, Y, test_size=0.12,random_state=0)
```

```python
print("X_train : ")
print(X_train*(np.max(X_before)-np.min(X_before))+np.min(X_before))
print("Y_train : ")
print(Y_train*(np.max(Y_before)-np.min(Y_before))+np.min(Y_before))
print("X_test : ")
print(X_test*(np.max(X_before)-np.min(X_before))+np.min(X_before))
print("Y_test : ")
print(Y_test*(np.max(Y_before)-np.min(Y_before))+np.min(Y_before))
```

```python
import tensorflow as tf

tf.set_random_seed(777)  # for reproducibility
learning_rate = 0.01

X = tf.placeholder(tf.float32, [None, 4])
Y = tf.placeholder(tf.float32, [None, ])

W1 = tf.Variable(tf.random_normal([4, 50]), name='weight1')
b1 = tf.Variable(tf.random_normal([50]), name='bias1')
layer1 = tf.nn.relu(tf.matmul(X, W1) + b1)

W2 = tf.Variable(tf.random_normal([50, 30]), name='weight2')
b2 = tf.Variable(tf.random_normal([30]), name='bias2')
layer2 = tf.nn.relu(tf.matmul(layer1, W2) + b2)

W3 = tf.Variable(tf.random_normal([30, 1]), name='weight3')
b3 = tf.Variable(tf.random_normal([1]), name='bias3')
hypothesis = tf.nn.sigmoid(tf.matmul(layer2, W3) + b3)

# cost/loss function
cost = -tf.reduce_mean(Y * tf.log(hypothesis) + (1 - Y) *
                       tf.log(1 - hypothesis))

train = tf.train.GradientDescentOptimizer(learning_rate=learning_rate).minimize(cost)

# Accuracy computation
# True if hypothesis>0.5 else False
# predicted = tf.cast(hypothesis > 0.5, dtype=tf.float32)
# accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, Y_test), dtype=tf.float32))

# Launch graph
with tf.Session() as sess:
    # Initialize TensorFlow variables
    sess.run(tf.global_variables_initializer())

    for step in range(20000):
        sess.run(train, feed_dict={X: X_train, Y: Y_train})
        if step % 100 == 0:
            print(step, sess.run(cost, feed_dict={
                  X: X_train, Y: Y_train})) #, sess.run([W1, W2]))

    # Accuracy report
    # h, c, a = sess.run([hypothesis, predicted, accuracy],
    #                   feed_dict={X: X_test, Y: Y_test})
    
    h = sess.run( hypothesis, feed_dict={X: X_test, Y: Y_test})
    
    print("\nHypothesis: ", h)
    #print( "\nCorrect: ", c)
    #print("\nAccuracy: ", a)
    print("\nrealY: ", Y_test)

    print("\nHypothesis: ")
    print(h*(np.max(Y_before)-np.min(Y_before))+np.min(Y_before))

    print("realY")
    print(Y_test*(np.max(Y_before)-np.min(Y_before))+np.min(Y_before))
    
    mse = mean_squared_error(Y_test, h)
    print("MSE: %.4f" % mse)
```
