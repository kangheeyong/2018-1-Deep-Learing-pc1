```python
learning_rate = 0.000001
tf.set_random_seed(777)

batchsize = 10

n_InputUnit = 3648
n_1stHiddenUnit = 500
n_2ndHiddenUnit = 200
n_3rdHiddenUnit = 60
n_OutputUnit = 3

K = 0.0
```

```python
# Algorithm which decides standard for classifying data
# Algorithm which can detect some kind of different phenomena

X = tf.placeholder(tf.float32, [None, n_InputUnit]) # tensorflow의 placeholder라는 변수를 선언
Y = tf.placeholder(tf.float32, [None, n_OutputUnit])
W = tf.placeholder(tf.float32, [None, n_InputUnit])
keep_prob = tf.placeholder(tf.float32)

W1 = tf.Variable(tf.random_normal([n_InputUnit, n_InputUnit],stddev=0.01))

W2 = tf.Variable(tf.random_normal([n_InputUnit, n_1stHiddenUnit],stddev=0.01))
W3 = tf.Variable(tf.random_normal([n_1stHiddenUnit, n_2ndHiddenUnit],stddev=0.01))
W4 = tf.Variable(tf.random_normal([n_2ndHiddenUnit, n_3rdHiddenUnit],stddev=0.01))
W5 = tf.Variable(tf.random_normal([n_3rdHiddenUnit, n_OutputUnit],stddev=0.01))

b2 = tf.Variable(tf.random_normal([n_1stHiddenUnit], stddev=0.01))
b3 = tf.Variable(tf.random_normal([n_2ndHiddenUnit], stddev=0.01))
b4 = tf.Variable(tf.random_normal([n_3rdHiddenUnit], stddev=0.01))
b5 = tf.Variable(tf.random_normal([n_OutputUnit], stddev=0.01))

L1 = tf.matmul(X, W1)

L2 = tf.nn.elu(tf.matmul(L1, W2) + b2)
L2_drop = tf.nn.dropout(L2, keep_prob)

L3 = tf.nn.elu(tf.matmul(L2_drop, W3) + b3)
L3_drop = tf.nn.dropout(L3, keep_prob)

L4 = tf.nn.elu(tf.matmul(L3_drop, W4) + b4)
logits = tf.matmul(L4, W5) + b5
Hypothesis = tf.nn.softmax(logits)

for i in range (3648):
    K = K + abs(W1[i,i])

#tf.stop_gradient(Y)
cost = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(labels=Y, logits=logits) + K)
#cost = cost = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(, Y))
train = tf.train.AdamOptimizer(learning_rate=learning_rate,beta1=0.5).minimize(cost)
prediction = tf.argmax(Hypothesis, 1)
correct_prediction = tf.equal(tf.argmax(Hypothesis, 1), tf.argmax(Y, 1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
```
