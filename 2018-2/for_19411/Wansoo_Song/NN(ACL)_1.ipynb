```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import statsmodels.formula.api as sm
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn import cross_validation
import tensorflow as tf
```

```python
data = np.genfromtxt('acl data.txt') # The default path is same with this ipynb

X_before = data[:,0:4] #The ":" means include whole range of array. 0:4 means to include numbers from 0th to 3rd column.
Y_before = data[:,4] #4 means to include numbers from 4th column.
X = (X_before-np.min(X_before))/(np.max(X_before)-np.min(X_before))
Y = (Y_before-min(Y_before))/(max(Y_before)-min(Y_before))   #nomalizing
# 배치 전체를 normalize? 하나의 수로?

dfX=pd.DataFrame(X) #Include X to dfX, and make it to table type.
dfy=pd.DataFrame(Y)

table=pd.concat([dfX, dfy], axis=1) #If axis = 1, then concat to right side. Otherwise (axis = 0), then concat to bottom side.
                                    #Also, if they have sharing indexes or columns, then it will be write on the same indexes or columns.
                                    #But, if it doesn't, then it will be write on the different indexes or columns, 
                                    #and it will be appear as NaN in the not shared parameters.
table.columns=['Power','Pressure','C3H6','N2','Thickness_Top'] #P

table
```

```python
X_train, X_test, Y_train, Y_test = cross_validation.train_test_split(
X, Y, test_size=0.14,random_state=0)
```

```python
print("X_train : ") #unnormalize
print(X_train*(np.max(X_before)-np.min(X_before))+np.min(X_before))
print("Y_train : ")
print(Y_train*(np.max(Y_before)-np.min(Y_before))+np.min(Y_before))
print("X_test : ")
print(X_test*(np.max(X_before)-np.min(X_before))+np.min(X_before))
print("Y_test : ")
print(Y_test*(np.max(Y_before)-np.min(Y_before))+np.min(Y_before))
```

```python
import tensorflow as tf

tf.set_random_seed(777)  # for reproducibility
learning_rate = 0.01

X = tf.placeholder(tf.float32, [None, 4])
Y = tf.placeholder(tf.float32, [None, ])

W1 = tf.Variable(tf.random_normal([4, 3]), name='weight1')
b1 = tf.Variable(tf.random_normal([3]), name='bias1')
layer1 = tf.nn.relu(tf.matmul(X, W1) + b1)

W2 = tf.Variable(tf.random_normal([3, 4]), name='weight2')
b2 = tf.Variable(tf.random_normal([4]), name='bias2')
layer2 = tf.nn.relu(tf.matmul(layer1, W2) + b2)

W3 = tf.Variable(tf.random_normal([4, 3]), name='weight3')
b3 = tf.Variable(tf.random_normal([3]), name='bias3')
layer3 = tf.nn.relu(tf.matmul(layer2, W3) + b3)

W4 = tf.Variable(tf.random_normal([3, 2]), name='weight4')
b4 = tf.Variable(tf.random_normal([2]), name='bias4')
layer4 = tf.nn.relu(tf.matmul(layer3, W4) + b4)

W5 = tf.Variable(tf.random_normal([2, 1]), name='weight5')
b5 = tf.Variable(tf.random_normal([1]), name='bias5')
hypothesis = tf.nn.sigmoid(tf.matmul(layer4, W5) + b5) #hypothesis is y_predict

# cost/loss function
# cost = -tf.reduce_mean(Y * tf.log(hypothesis) + (1 - Y) * tf.log(1 - hypothesis))<-Logistic
cost = tf.reduce_mean(tf.square(hypothesis - Y)) # <-MSE

train = tf.train.GradientDescentOptimizer(learning_rate=learning_rate).minimize(cost)
                #AdamOptimizer(...)...

# Accuracy computation
# True if hypothesis>0.5 else False
# predicted = tf.cast(hypothesis > 0.5, dtype=tf.float32)
# accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, Y_test), dtype=tf.float32))

# Launch graph
with tf.Session() as sess:
    # Initialize TensorFlow variables
    sess.run(tf.global_variables_initializer())

    for step in range(20000):
        sess.run(train, feed_dict={X: X_train, Y: Y_train})
        if step % 100 == 0:
            print(step, sess.run(cost, feed_dict={
                  X: X_train, Y: Y_train})) #, sess.run([W1, W2]))

    # Accuracy report
    # h, c, a = sess.run([hypothesis, predicted, accuracy],
    #                   feed_dict={X: X_test, Y: Y_test})
    
    h = sess.run( hypothesis, feed_dict={X: X_test, Y: Y_test})
    
    print("\nHypothesis: ", h)
    #print( "\nCorrect: ", c)
    #print("\nAccuracy: ", a)
    print("\nrealY: ", Y_test)

    print("\nHypothesis: ")
    print(h*(np.max(Y_before)-np.min(Y_before))+np.min(Y_before))

    print("realY")
    print(Y_test*(np.max(Y_before)-np.min(Y_before))+np.min(Y_before))
    
    mse = mean_squared_error(Y_test, h)
    print("MSE: %.4f" % mse)

    a = r2_score(Y_test,h)
    print("R2: %.4f" % a)
```
