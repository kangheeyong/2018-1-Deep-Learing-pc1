```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn import datasets
```

## Used Dataset
### Iris Data

종속변수(dependent variable) : Species

독립변수(independent variable) :Sepal.Length,Sepal.Width, Petal.Length, Petal.Width

```python
iris = datasets.load_iris()
dfx = pd.DataFrame(iris.data) 
dfy = pd.DataFrame(iris.target)
dfiris=pd.concat([dfx,dfy],axis=1)
dfiris.columns=['Sepal.Length','Sepal.Width','Petal.Length','Petal.Width','Species'] # 0=setosa / 1=versicolor / 2=virginica
dfiris['Species'] = dfiris['Species'].map({0: "setosa", 1: "versicolor", 2: "virginica"})
dfiris
```

```python
sns.pairplot(dfiris, hue='Species')
```

## Split dataset

```python
from sklearn.model_selection import train_test_split

x_train, x_test , y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3, stratify=iris.target, random_state=123456)
```

## RandomForest

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=10, oob_score=True, random_state=123456) #oob 예측이 얼마나 정확한가에 대한 추정을 수치로 나타냄
rf=rf.fit(x_train, y_train)
```

## Result

```python
rf.feature_importances_ 
```

```python
from sklearn.metrics import accuracy_score


predicted = rf.predict(x_test)
accuracy = accuracy_score(y_test, predicted)

print(predicted)
print(accuracy)
print(f'Out-of-bag score estimate: {rf.oob_score_:.3}')
print(f'Mean accuracy score: {accuracy:.3}')
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, rf.predict(x_test)))
```

```python
rf_y_pred=rf.predict(x_test)
rf_y_pred
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, rf_y_pred)
```

```python


cm = pd.DataFrame(confusion_matrix(y_test, predicted), columns=iris.target_names, index=iris.target_names)
sns.heatmap(cm, annot=True)
```

# Bagging + Decision tree

```python
from sklearn.model_selection import train_test_split

x_train, x_test , y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2, stratify=iris.target, random_state=123456)
```

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier


model2 = BaggingClassifier(DecisionTreeClassifier(), bootstrap_features=True, random_state=123456,n_estimators=100).fit(x_train, y_train)

```

```python
from sklearn.metrics import accuracy_score

predicted2 = model2.predict(x_test)
accuracy2 = accuracy_score(y_test, predicted2)

print(f'Mean accuracy score: {accuracy2:.3}')
```

```python
confusion_matrix(y_test, predicted2)
```

```python
from sklearn.metrics import confusion_matrix

cm = pd.DataFrame(confusion_matrix(y_test, predicted2), columns=iris.target_names, index=iris.target_names)
sns.heatmap(cm, annot=True)
```
