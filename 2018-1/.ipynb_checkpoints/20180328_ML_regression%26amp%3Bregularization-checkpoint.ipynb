numpy(numerical python) : 선형대수(배열,행렬) 연산에 효과적인 라이브러리

          An effective library for linear algebra(arrays, matrices)

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import statsmodels.formula.api as sm
from sklearn.model_selection import train_test_split
```

## Used Dataset
### 당뇨병 환자 442명의 검사 데이터 from sklearn  (442 diabetic patients data from sklearn)
종속변수(dependent variable) : 1년 후의 당뇨병 진행도 (diabetes progression after 1 year)

독립변수(independent variable) : 나이, 성별, BMI지수, 혈압 등 10개의 변수 (age, sex, BMI index, pressure pressure, etc.)

```python
diabetes = datasets.load_diabetes()
dfX=pd.DataFrame(diabetes.data)
dfy=pd.DataFrame(diabetes.target)
dfdiabetes=pd.concat([dfX, dfy], axis=1)
dfdiabetes.columns=['X1','X2','X3','X4','X5','X6','X7','X8','X9','X10','y']
dfdiabetes
```


### Split dataset

```python
print(len(diabetes.data)*0.7)
X_train, X_test , y_train, y_test = train_test_split(diabetes.data, diabetes.target, test_size=0.3, random_state=1)
dfXtrain=pd.DataFrame(X_train)
dfytrain=pd.DataFrame(y_train)
dfXtest=pd.DataFrame(X_test)
dfytest=pd.DataFrame(y_test)

dftrain=pd.concat([dfXtrain,dfytrain],axis=1)
dftest=pd.concat([dfXtest,dfytest],axis=1)
dftrain.columns=['X1','X2','X3','X4','X5','X6','X7','X8','X9','X10','y']
dftest.columns=['X1','X2','X3','X4','X5','X6','X7','X8','X9','X10','y']
dftrain
```

## Linear Regression

http://scikit-learn.org/stable/modules/linear_model.html

### create model and make prediction

```python
LinReg = linear_model.LinearRegression()
LinReg.fit(X_train[:,2:3], y_train)
LinReg_y_pred = LinReg.predict(X_test[:,2:3])
LinReg_y_pred
```

### plotting

```python
plt.scatter(X_test[:,2:3], y_test,  color='black')
plt.plot(X_test[:,2:3], LinReg_y_pred, color='blue', linewidth=3)
plt.show()
```

### result

```python
print('Coefficients:', LinReg.coef_)
print('MeanSquaredError: %.2f'
     % mean_squared_error(y_test, LinReg_y_pred))
print('MeanAbsoluteError: %.2f'
     % mean_absolute_error(y_test, LinReg_y_pred))
print('R Square: %.2f'
     % r2_score(y_test, LinReg_y_pred), '\n')
```

## Multivariate Regression using sklearn package

### create model and make prediction

```python
MulReg=linear_model.LinearRegression()
MulReg.fit(X_train, y_train)

MulReg_y_pred = MulReg.predict(X_test)
MulReg_y_pred
```

### result

```python
print('Coefficients:', MulReg.coef_)
print('MeanSquaredError: %.2f'
     % mean_squared_error(y_test, MulReg_y_pred))
print('MeanAbsoluteError: %.2f'
     % mean_absolute_error(y_test, MulReg_y_pred))
print('R Square: %.2f'
     % r2_score(y_test, MulReg_y_pred), '\n')
```

## Multivariate Regression using statsmodels packages

http://www.statsmodels.org/dev/generated/statsmodels.regression.linear_model.OLS.html

### create model and summary

```python
MulReg2 = sm.ols(formula='y ~ X1+X2+X3+X4+X5+X6+X7+X8+X9+X10', data=dftrain)  # OLS : Ordinary Least Square Model
MulReg2 = MulReg2.fit()
MulReg2.summary()
```

### make prediction

```python
MulReg2.params
MulReg2_y_pred = MulReg2.predict(dftest)
MulReg2_y_pred
```

### result

```python
#print('Coefficients:', MulReg.coef_)
print('MeanSquaredError: %.2f'
     % mean_squared_error(y_test, MulReg2_y_pred))
print('MeanAbsoluteError: %.2f'
     % mean_absolute_error(y_test, MulReg2_y_pred))
print('R Square: %.2f'
     % r2_score(y_test, MulReg2_y_pred), '\n')
```

## Ridge Regression (Linear least squares with l2 regularization)

http://scikit-learn.org/stable/modules/generated/sklearn.linear_model.Ridge.html

### create model and make prediction

```python
ridge = linear_model.Ridge(alpha=0.35)
ridge.fit(X_train, y_train)
ridge_y_pred = ridge.predict(X_test)
ridge_y_pred
```

### result

```python
print('Coefficients:', ridge.coef_)
print('MeanSquaredError: %.2f'
     % mean_squared_error(y_test, ridge_y_pred))
print('MeanAbsoluteError: %.2f'
     % mean_absolute_error(y_test, ridge_y_pred))
print('R Square: %.2f'
     % r2_score(y_test, ridge_y_pred), '\n')
```

## Lasso Regression (Linear least squares with l1 regularization)

### create model and make prediction

```python
lasso = linear_model.Lasso(alpha=0.35)
lasso.fit(X_train, y_train)
lasso_y_pred = lasso.predict(X_test)
lasso_y_pred
```

### result

```python
print('Coefficients:', ridge.coef_)
print('MeanSquaredError: %.2f'
     % mean_squared_error(y_test, lasso_y_pred))
print('MeanAbsoluteError: %.2f'
     % mean_absolute_error(y_test, lasso_y_pred))
print('R Square: %.2f'
     % r2_score(y_test, lasso_y_pred), '\n')
```
